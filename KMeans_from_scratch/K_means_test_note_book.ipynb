```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import KMeans3 as km3
import KMeans5 as km5
```

## **Dataset generer**

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=100, n_features=2, centers = 2, random_state= 40)
```

#### Premier modele: modele avec kmeans++

```python
model = km5.Kmeans(2)
model.fit(X)
plt.scatter(X[:,0],X[:,1], c = model.predict(X))
plt.scatter(model.centroid[:,0],model.centroid[:,1], c = 'r')
plt.title("Agregation de mon modele (cas de presence d'initialisation k-means++)")
```

```python
label, centroid = model.optimiser(20)
```

```python
#model = km5.Kmeans(2)
#model.fit(X)
plt.scatter(X[:,0],X[:,1], c = label)
plt.scatter(centroid[:,0],centroid[:,1], c = 'r')
plt.title("Agregation de mon modele (cas de presence d'initialisation k-means++)")
```

#### modele sans kmeans++

```python
model = km3.Kmeans(2)
model.fit(X)
plt.scatter(X[:,0],X[:,1], c = model.predict(X))
plt.scatter(model.centroid[:,0],model.centroid[:,1], c = 'r')
plt.title("Agregation de mon modele (cas d'absence d'initialisation k-means++)")
```

#### K-means de SciKit Learn

```python
skmodel = KMeans(n_clusters = 2)
skmodel.fit(X)
plt.scatter(X[:,0],X[:,1], c = skmodel.predict(X) )
plt.scatter(skmodel.cluster_centers_[:, 0],skmodel.cluster_centers_[:, 1],c= 'r')
plt.title("Agregation Kmeans avec SciKit Learn")
```
